```python
def dataCleaning(data,column):
  data.select(column).distinct().show(10,False)
  null_values_count = data.select(column).filter(col(column).isNull()).count()
  print("null_count ->",null_values_count)
  print("characters_count ->",data.select(column).filter(col(column).rlike("^[A-Za-z]+")).distinct().count())
  print("integer_count ->",data.select(column).filter(~col(column).rlike("^[A-Za-z]+")).distinct().count())
  print("null values percentage ->",null_values_count/data.count()*100)
  print("column_dataType ->",data.schema[column].dataType)
```
